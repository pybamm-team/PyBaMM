```python
%autoreload 2
model = pybamm.lithium_ion.DFN({"SEI": "my sei"})

params = pybamm.ParameterValues("OKane2022")
experiment = pybamm.Experiment(
    [
        (
            "Discharge at 1C until 3V",
            "Rest for 1 hour",
            "Charge at 1C until 4.1V",
        )
    ]
    * 50
)

sim = pybamm.Simulation(model, experiment=experiment, parameter_values=params)

sol = sim.solve()
```
