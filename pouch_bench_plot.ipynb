```python
import matplotlib.pyplot as plt 
import pandas as pd
```

## Load data

```python
df = pd.read_csv("pouch_bench.csv")

npts = df["N"].to_numpy()
set_up_times = df["Setup time"].to_numpy()
solve_times = df["Solve time"].to_numpy()
integration_times = df["Integration time"].to_numpy()
states = df["States"].to_numpy()
```

## Plot

```python
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(npts, set_up_times, "-o")
ax[0, 0].set_xlabel("Number of coupled DFN models")
ax[0, 0].set_ylabel("Set up time [s]")
ax[0, 1].plot(npts, solve_times, "-o")
ax[0, 1].set_xlabel("Number of coupled DFN models")
ax[0, 1].set_ylabel("Solve time [s]")
ax[1, 0].plot(npts, integration_times, "-o")
ax[1, 0].set_xlabel("Number of coupled DFN models")
ax[1, 0].set_ylabel("Integration time [s]")
ax[1, 1].loglog(states, integration_times, "-o")
ax[1, 1].set_xlabel("Number of states")
ax[1, 1].set_ylabel("Integration time [s]")
plt.show()
```
